```python
import pandas as pd
import numpy as np
import math



df = pd.read_csv('/content/drive/MyDrive/S2S/hindi/merged_chunks/merged_batch_0.csv')
total_rows = len(df)
# -----------------------------------------------------
max_chunk_size = 300

# 1. Calculate the number of chunks and remainder
num_chunks = math.ceil(total_rows / max_chunk_size)
base_size = total_rows // num_chunks
remainder = total_rows % num_chunks

# 2. Create the list of chunk sizes
chunk_sizes = (
    [base_size + 1] * remainder +
    [base_size] * (num_chunks - remainder)
)

# 3. Use cumulative sum to find the slicing indices
indices = np.cumsum([0] + chunk_sizes)

# 4. Perform the slicing and store DataFrames in a list
chunked_dfs = []
for i in range(num_chunks):
    start = indices[i]
    end = indices[i+1]
    chunk = df.iloc[start:end]
    chunked_dfs.append(chunk)

print(f"Divided into {num_chunks} DataFrames.")
print(f"Total rows processed: {sum(len(d) for d in chunked_dfs)}")
print(f"First chunk size: {len(chunked_dfs[0])}")
print(f"Last chunk size: {len(chunked_dfs[-1])}")
```

```python
#gtts
from gtts import gTTS
#en-IN-NeerjaNeural,en-IN-PrabhatNeural
import edge_tts
import os
import asyncio
#couqi
import pandas as pd
from TTS.api import TTS
```

```python
output_dir = ''
os.makedirs(output_dir, exist_ok=True)

for k in range(0,0):
  tts = TTS(model_name="tts_models/en/ljspeech/tacotron2-DDC", progress_bar=False, gpu=False)

  df_chunk = chunked_dfs[k]

  for i, row in df_chunk.iterrows():
      text = str(row['en_text']).strip()
      if pd.notna(text) and text != "":
          filename = os.path.join(output_dir, f"audio_0_{i}.wav")
          tts.tts_to_file(text=text, file_path=filename)

          print(f"Generated: {filename}")


  print(f"\n Coqui TTS conversion complete for chunk {k}!\n")
  print()
  #---------------------------------------------------------------

  VOICE = "en-IN-PrabhatNeural"

  async def main():
      for i, row in chunked_dfs[k+1].iterrows():
          text = str(row["en_text"])
          if pd.notna(text) and text.strip() != "":
              filename = os.path.join(output_dir, f"audio_0_{i}.wav")
              communicate = edge_tts.Communicate(text, voice=VOICE)
              await communicate.save(filename)

              print(f"Generated: {filename}")


      print(f"CSV to Audio conversion complete with Edge-TTS dfchunk[{k+1}]!")
      print()

  #Run it ---
  await main()

  #--------------------------------------------------------------------

  VOICE = "en-IN-NeerjaNeural"

  async def main():
      for i, row in chunked_dfs[k+2].iterrows():
          text = str(row["en_text"])
          if pd.notna(text) and text.strip() != "":
              filename = os.path.join(output_dir, f"audio_0_{i}.wav")
              communicate = edge_tts.Communicate(text, voice=VOICE)
              await communicate.save(filename)

              print(f"Generated: {filename}")


      print(f"CSV to Audio conversion complete with Edge-TTS dfchunk[{k+2}]!")
      print()

  #Run it ---
  await main()

  #---------------------------------

  for i,row in chunked_dfs[k+3].iterrows():
      text = str(row['en_text'])
      if pd.notna(text) and text.strip()!=" ":
          tts=gTTS(text=text,lang='en',tld='co.in')
          filename = os.path.join(output_dir,f"audio_0_{i}.wav")
          tts.save(filename)
          print(f"Generated: {filename}")

  print(f"gtts CSV to Audio conversion complete for dfchunk[{k+3}]!")
  print()

  #------------------------------

  for i,row in chunked_dfs[k+4].iterrows():
      text = str(row['en_text'])
      if pd.notna(text) and text.strip()!=" ":
          tts=gTTS(text=text,lang='en',tld='co.in')
          filename = os.path.join(output_dir,f"audio_0_{i}.wav")
          tts.save(filename)
          print(f"Generated: {filename}")
  print()
  print(f"gtts CSV to Audio conversion complete for dfchunk[{k+4}]!")
```
